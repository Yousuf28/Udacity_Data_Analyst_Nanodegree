### HAVING

```python
# # SELECT COLUMN_1, AGGREGATE_FUNCTION/OR NOT (COLUMN_2)
# FROM TABLE_NAME GROUP BY COLUMN_1
# HAVING condition
```

```python
pd.read_sql(' SELECT customer_id, SUM(amount) FROM payment GROUP BY customer_id limit 10' , engine)
```

```python
pd.read_sql(' SELECT customer_id, SUM(amount) FROM payment GROUP BY customer_id HAVING SUM(amount) > 200' , engine)
```

```python
pd.read_sql('SELECT store_id, COUNT(customer_id) FROM customer GROUP BY store_id limit 5' , engine)
```

```python
pd.read_sql('SELECT store_id, COUNT(customer_id) FROM customer GROUP BY store_id HAVING COUNT(customer_id)>300' , engine)
```

```python
pd.read_sql("SELECT rating, AVG(rental_rate) FROM film WHERE rating IN ('R', 'G', 'PG') GROUP BY rating HAVING AVG(rental_rate)<3" , engine)
```
